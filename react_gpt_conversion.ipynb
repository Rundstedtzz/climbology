```python
import json

holds_list = [{'id': 443, 'x_grid': 1, 'y_grid': 35, 'x_coor': 368, 'y_coor': 31, 'hold_type': 'jug', 'hold_function': 'both', 'hold_depth': 2, 'hold_orientation': '0', 'isSelected': False, 'hold_position_in_route': 'Start'}, {'id': 328, 'x_grid': 13, 'y_grid': 25, 'x_coor': 624, 'y_coor': 247, 'hold_type': 'pinch', 'hold_function': 'both', 'hold_depth': 1, 'hold_orientation': '15', 'isSelected': False, 'hold_position_in_route': 'End'}, {'id': 250, 'x_grid': 13, 'y_grid': 19, 'x_coor': 621, 'y_coor': 374, 'hold_type': 'jug', 'hold_function': 'both', 'hold_depth': 2, 'hold_orientation': '0', 'isSelected': False, 'hold_position_in_route': 'Start'}, {'id': 226, 'x_grid': 17, 'y_grid': 17, 'x_coor': 712, 'y_coor': 416, 'hold_type': 'jug', 'hold_function': 'both', 'hold_depth': 3, 'hold_orientation': '180', 'isSelected': False, 'hold_position_in_route': 'Middle'}, {'id': 230, 'x_grid': 25, 'y_grid': 17, 'x_coor': 879, 'y_coor': 422, 'hold_type': 'jug', 'hold_function': 'both', 'hold_depth': 2, 'hold_orientation': '0', 'isSelected': False, 'hold_position_in_route': 'Foot Only'}, {'id': 334, 'x_grid': 25, 'y_grid': 25, 'x_coor': 914, 'y_coor': 249, 'hold_type': 'crimp', 'hold_function': 'both', 'hold_depth': 2, 'hold_orientation': '-15', 'isSelected': False, 'hold_position_in_route': 'Start'}, {'id': 457, 'x_grid': 29, 'y_grid': 35, 'x_coor': 969, 'y_coor': 35, 'hold_type': 'jug', 'hold_function': 'both', 'hold_depth': 3, 'hold_orientation': '60', 'isSelected': False, 'hold_position_in_route': 'End'}]

def convert_holds_to_json(holds_list):
    converted_holds = []

    for hold in holds_list:
        if isinstance(hold, dict):
            converted_hold = {
                "id": hold.get('id'),
                "x_coordinate": hold.get('x_grid'),
                "y_coordinate": hold.get('y_grid'),
                "type": hold.get('hold_type'),
                "depth": hold.get('hold_depth'),
                "orientation": hold.get('hold_orientation'),
                "position_in_route": hold.get('hold_position_in_route', 'middle') if hold.get('hold_position_in_route') != "Foot Only" else "middle",
                "foot_restriction": "foot-only" if hold.get('hold_position_in_route') == "Foot Only" else "foot-or-hand"
            }
            converted_holds.append(converted_hold)

    return {"holds": converted_holds}

# Convert the holds list to JSON format
converted_json = convert_holds_to_json(holds_list)
print(json.dumps(converted_json, indent=4))
```
